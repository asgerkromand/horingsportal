```python
from pathlib import Path
import json
import pandas as pd
```

```python
fp_project = Path("/Users/asgerkromand/hoering")
fp_output = fp_project / "output" / "test_sets"
fp_data = fp_project / "data" / "hearings"
```

```python
# List of follow-up cases
fp_cases = [
    64359,
    68308,
    16958,
    68987,
    16626,
    61474,
    65637,
    65029,
    64606,
    66443,
    68730,
    60352,
    63925,
    61809,
    67083
]
```

```python
# Load in .json files in specified directory

cases = []
for case in fp_cases:
    case_number = str(case)
    fp_case = fp_data / case_number
    # Load in the .json file
    with open(fp_case / f"{case}_meta.json", "r") as f:
        data = json.load(f)

    # Extract the text from the JSON data
    case_data = {
        "høringsnummer": case,
        "officiel titel": data['Officiel titel'] if 'Officiel titel' in data else None,
        "beskrivelse": data['Beskrivelse'] if 'Beskrivelse' in data else None,
        "høringstype": data['Høringstype'] if 'Høringstype' in data else None,
        "høringsfrist": data['Høringsfrist'] if 'Høringsfrist' in data else None,
        "myndighed": data['Myndighed'] if 'Myndighed' in data else None,
        "område": data['Område'] if 'Område' in data else None,
        "høringsfrist": data['Høringsfrist'] if 'Høringsfrist' in data else None,
        "arkiveringsdato": data['Arkiveringsdato'] if 'Arkiveringsdato' in data else None,
        "ikræfttrædelsesdato": data['Ikrafttrædelsesdato'] if 'Ikrafttrædelsesdato' in data else None,
        "høringsår": data['Høringsår'] if 'Høringsår' in data else None,
        "kontaktperson": data['Kontaktperson'] if 'Kontaktperson' in data else None,
        "kontaktperson_e-mail": data['Kontaktperson e-mail'] if 'Kontaktperson e-mail' in data else None,
        "publiceringsdato": data['Publiceringsdato'] if 'Publiceringsdato' in data else None,
    }
    # Append the case data to the list
    cases.append(case_data)

# Create a DataFrame from the list of dicts
df_cases = pd.DataFrame(cases)
```

```python
# Sort by myndighed and høringsår
df_cases = df_cases.sort_values(by=['myndighed', 'høringsår'])
df_cases
```
